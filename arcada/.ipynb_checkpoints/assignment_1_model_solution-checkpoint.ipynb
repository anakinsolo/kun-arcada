## Median (aka expected value)

You'd like to have a function that takes an unsorted pandas.Series object (or a list) and returns a median for it.

Sort the serie: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.sort_values.html (or your list).

There are two opotions: odd and even number of elements.

If the number is odd, return an element with the index 'length of serie floor divided by 2'

If the number is even, return a half sum of two middle elements.

You can use `.iloc[]` to access the position of a row if pandas.Series object. Remember that the position is zero-based. https://stackoverflow.com/a/31593712

```python
# model
def median (serie:pd.Series):
    if not isinstance(serie, pd.Series):
        raise Exception("Sorry, the function accepts a pandas Series as argument")
    ser = serie.copy().dropna().sort_values()
    i = ser.size//2
    return ser.iloc[i] if len(ser)%2 else ser.iloc[i-1:i+1].sum()/2
```

```python
# prety printing function
# you should provide function's name as argument to parameter 'fun'
columns = ['temp_avg', 'light_avg', 'humidity_avg']
def print_columns(fun, group, indent=0, c=columns):
    for col in c:
        tab = '\t'*indent
        key = col.replace('_avg', '')
        value = fun(group[col])
        print(f'{tab}{key:<10} {value:>5.2f}')
```

```python
# prety printing function
# you should provide function's name as argument to parameter 'fun'
def print_stats(fun, title='',  data=None):
    print (title)
    print("="*40)
    for loc_id, loc_data in data.groupby('location'):
        print(loc_id)
        print_columns(fun, loc_data, indent=1)
        print()
        for board_id, board_data in loc_data.groupby('boardid'):
            elevation = board_data["elevation"].median()
            count = board_data["boardid"].count()
            print(f'\tboard {board_id} @ elevation {elevation:.1f} ({count} data points)')
            print_columns(fun, board_data, indent=2)
        print("="*40)
```

## Standard deviation

Write a function that takes pandas.Series object and returns standard deviation.

First calculate `mean = serie.sum() / serie.count()`

Then make a Series object which contains `squares_off_difference` between every measured value and the `mean`.

Then return sqrt of ratio of `squares_off_difference.sum()` to number of elements in `serie` minus 1.

```python
# model
# .size includes NaN values, .count() does not:
def st_dev (ser:pd.Series):
    n = ser.count()
    mean = ser.sum()/n
    sum_sq_diff = ((ser - mean)**2).sum()
    return (sum_sq_diff/(n-1))**.5
```

```python
test_data = list(range(10))+[None]
display(test_data)
test = pd.Series(test_data)
print(" my:", st_dev(test));  print("lib:", test.std())
```

```python
# helper function
# see https://matplotlib.org/api/axes_api.html
def get_sub_ax(ax, xbound=(data.index.min(), data.index.max()), ybound=(0, 100)):
    ax_i = ax.twinx(); # https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.twinx.html
    ax_i.set_ybound(ybound); # https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_ybound.html
    ax_i.set_xbound(xbound)
    ax_i.set_autoscaley_on(False); 
    ax_i.yaxis.tick_left()
    ax_i.grid(b=True, color='b', alpha=.1, axis='y') # https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.grid.html
    return ax_i
```

```python
# automation ploting with a function
def plot_time_range (t_min_str, t_max_str, marker=',', resample='15Min'):
    # plt (=matplotlib.pyplot) is just a state-based INTERFACE to matplotlib 
    # se https://matplotlib.org/api/pyplot_summary.html
    
    # basic preparations
    fig, ax = plt.subplots(figsize=(18,7))
    ax.set_yticks([]) # remove from empty axes, we will not plot there 
    ax.grid(b=True, color='b', alpha=.1, axis='x')
    
    # for color separation of 2*3 series
    color_groups = (list('rgb'), list('myc')) 
    
    t_min = pd.Timestamp(t_min_str)
    t_max = pd.Timestamp(t_max_str)
    
    for (loc_id, data_by_loc), colors, legend_yshift \
    in zip(data.groupby('location'), color_groups, [.6,.3]):
        # take a time slice of data
        data_time_slice = data_by_loc[
            (data_by_loc.index > t_min) & (data_by_loc.index < t_max)
        # get rid of exessive points: http://pandas.pydata.org/pandas-docs/stable/timeseries.html#resampling
        ].resample(resample).median() 
        
        # make a twin to the base Axes (a fix for a problem(bug?) with colors in legends)
        ax_i = get_sub_ax(ax, 
                          xbound=(t_min, t_max),
                          ybound=(
                              data[measurments].values.min(), 
                              data[measurments].values.max()*1.1) # some space above
                         )
        handels = []
        # plot columns for this location with predefined colors
        # measurments = ['temp_avg', 'light_avg', 'humidity_avg']
        for measurment, color in zip(measurments, colors):
            # see https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html  > Notes
            ax_i.plot(
                data_time_slice.index, 
                data_time_slice[measurment], 
                marker, # https://matplotlib.org/api/markers_api.html#module-matplotlib.markers
                c=color)
            
            # https://matplotlib.org/gallery/text_labels_and_annotations/custom_legends.html
            handels.append(Line2D([0], [0], marker='o', color='w', label=measurment,
                          markerfacecolor=color, markersize=8))
        
        # config legend for this location
        if marker == ',': # difficult to see small markers, make them custom
            ax_i.legend(handels, measurments, loc=3, title=loc_id, bbox_to_anchor=(1, legend_yshift))
        else:
            ax_i.legend(measurments, markerscale=2, loc=3, title=loc_id, bbox_to_anchor=(1, legend_yshift))
    
    # display
    plt.show()
```
